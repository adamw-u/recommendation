```python
def stackModel(clf,NFOLDS=5):

    ntrain = x_train.shape[0]
    ntest = x_test.shape[0]
    
    kf = KFold(ntrain, n_folds=NFOLDS, random_state=0)

    oo_train = np.zeros((ntrain,2))
    oo_test = np.zeros((ntest,2))
    print "model %s"%(clf)
    print "--------------------"
    for i, (train_index, test_index) in enumerate(kf):
        x_tr = x_train[train_index]
        y_tr = y_train[train_index]
        x_te = x_train[test_index]
        y_te = y_train[test_index]

        clf.fit(x_tr, y_tr)

        oo_train[test_index] = clf.predict_proba(x_te)
        oo_test += clf.predict_proba(x_test)
        print "folds %d precision is %f"%(i,precision_score(y_te,clf.predict(x_te)))

    oo_test /= NFOLDS
    return oo_train, oo_test
```
